## Описание проекта и постановка задачи

У вас есть файлы с результатами запросов из предыдущих заданий.

**query_1.csv** — результат первого запроса. В нём содержится такая информация:

- *model* — модель самолета;
- *flights_amount* — количество рейсов для каждой модели самолетов *model* в сентябре 2018 года.

**query_3.csv** — результат третьего запроса. В нём содержится такая информация:

- *city* — город;
- *average_flights* — среднее количество рейсов, прибывающих в город (*city*) за день в сентябре 2018 года.

 Для этих двух наборов данных нужно:

- импортировать файлы;
- изучить данные в них;
- проверить типы данных на корректность;
- выбрать топ-10 городов по количеству рейсов;
- построить графики: модели самолётов и количество рейсов, города и количество рейсов, топ-10 городов и количество рейсов;
- сделать выводы по каждому из графиков, пояснить результат.

<a name="2"></a>
## Изучение и подготовка данных

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display
```

```python
query_1 = pd.read_csv('query_1.csv')
query_1.info()
display(query_1)
query_1.describe()
```

```python
query_3 = pd.read_csv('query_3.csv')
query_3.info()
display(query_3.sample(5))
query_3.describe()
```

```python
query_3['city'].nunique()
```

Типы данных корректны.

<a name="3"></a>
## Анализ

<a name="31"></a>
### Топ-10 городов по количеству рейсов

```python
top_cityes = query_3.sort_values(by='average_flights', ascending = False).head(10)
```

```python
display(top_cityes)
```

<a name="32"></a>
### График "модели самолётов и количество рейсов"

```python
plt.figure(figsize=(10, 4))
sns.set_style('whitegrid')
ax = sns.barplot(y='model', x='flights_amount', data=query_1)
plt.title('Модели самолётов и количество рейсов', fontsize=20)
plt.ylabel('Модель', fontsize=16)
plt.xlabel('Количество рейсов', fontsize=16);
```

Модель | Кол-во посадочных мест
--- |--- 
airbus a319-100 | 124-156
airbus a321-200 | 170-220
boeing 737-300  | 269-328
boeing 767-300 | 170-220
boeing 777-300 | 386-550
bombardier crj-200  | 40-50
cessna 208 caravan | 9-13
sukhoi superjet-100 | 87-108


Небольшие самолеты (bombardier crj-200, cessna 208 caravan, sukhoi superjet-100) совершили более 4000 рейсов (каждый). При этом крупные, задействованные на дальних маршрутах (боинги и аэробусы) менее 1000.

<a name="33"></a>
### Графики "города и количество рейсов"

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x='city', y='average_flights', data=query_3)

plt.xticks(rotation=90)
plt.title('Города и количество рейсов', fontsize=20)
plt.ylabel('Количество рейсов', fontsize=16)
plt.xlabel('Город', fontsize=16);
```

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x='city', y="average_flights", data=query_3.sort_values(by='average_flights', ascending = False))

plt.xticks(rotation=90)
plt.title('Города и количество рейсов (по убыванию)', fontsize=20)
plt.ylabel('Количество рейсов', fontsize=16)
plt.xlabel('Город', fontsize=16);
```

```python
plt.figure(figsize=(10, 4))
ax = sns.barplot(y='city', x="average_flights", data=top_cityes)

plt.title('Топ-10 городов по количеству рейсов', fontsize=20)
plt.xlabel('Количество рейсов', fontsize=16)
plt.ylabel('Город', fontsize=16);
```

```python
query_3['average_flights'].sum()
```

```python
top_cityes['average_flights'].sum()
```

Почти половина всех рейсов приходится на 10 лидеров (а на Москву почти четверть).
